Пользователь вводит целое число, программа складывает все цифры числа, с полученным числом — то же самое, и так до тех пор, пока не получится однозначное число.
545 -> 5
12345 -> 6

```python
#num = int(input("type your number:"))
num = 12345

while not num // 10 == 0:
    # get sum of digits of number 
    sum = 0
    while num:
        sum += num % 10
        num = num // 10
    num = sum
print(num)
```

Кинотеатр. Дан список списков, каждый вложенный список состоит из 1 и 0, количество вложенных списков — количество рядов. Пользователь вводит, сколько билетов ему требуется. Программа должна найти ряд, где можно приобрести нужное количество билетов (места должны быть рядом). Если таких рядов несколько, то ближайший к экрану (ближайшим считается нулевой ряд). Если таких мест нет, то вывести False.
[[0,1,1,0], [1,0,0,0], [0,1,0,0]], 2 -> 1
[[0,1,1,0], [1,0,1,0], [1,1,0,1]], 2 -> False


```python
def if_have_place(seats, tickets):
    for row in seats:
        free_tickets = 0
        for s in row:
            if s == 0:
                free_tickets += 1
            else:
                free_tickets = 0
            if free_tickets == tickets:
                return True
    return False
seats_1 = [[0,1,1,0], [1,0,0,0], [0,1,0,0]]
seats_2 = [[0,1,1,0], [1,0,1,0], [1,1,0,1]]
seats_3 = [[0,1,1,0], [1,0,1,0], [1,1,0,1]]
tickets = 4
print(if_have_place(seats_1, tickets))
        
```

Написать упрощенную версию алгоритма RLE. Алгоритм RLE объединяет подряд идущие символы в коэффициент и символ.

Пример:

aaabbbbccccc -> 3a4b5c
asssdddsssddd -> 1a3s3d3s3d
abcba -> 1a1b1c1b1a

```python
def RLE(text):
    combined_text = ""
    counter = 1
    current_symbol = text[0]
    for symbol in text[1:]:
        if symbol == current_symbol:
            counter += 1
        else:
            combined_text += str(counter) + current_symbol
            counter = 1
            current_symbol = symbol
    combined_text += str(counter)
    combined_text += current_symbol
    return combined_text


text_1 = "aaabbbbccccc"
text_2 = "asssdddsssddd"
text_3 = "abcba"
print(RLE(text_1))
print(RLE(text_2))
print(RLE(text_3))
```

Шифр Цезаря. Пользователь вводит строку и ключ шифра, программа должна вывести зашифрованную строку (со сдвигом по ключу). Сдвиг циклический. Используем только латинский алфавит, пробелы не шифруются.

Пример:

Dog, 2 -> Fqi
Zak zak -> Cdn cdn
Python is the BEST -> UdymtsnxymjGJXY


```python
def caesar_cipher(text, key):
    result = ""
    
    for char in text:
        # Для маленьких букв
        if char.islower():
            result += chr(ord('a') + (ord(char) - ord('a') + key) % 26)
        # Для больших букв
        elif char.isupper():
            result += chr(ord('A') + (ord(char) - ord('A') + key) % 26)
        # Пробелы и другие символы остаются без изменений
        else:
            result += char
    
    return result

print(caesar_cipher("Dog", 2)) 
print(caesar_cipher("Zak zak", 2)) 
print(caesar_cipher("Python is the BEST", 2)) 
```
