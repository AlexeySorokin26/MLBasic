```python
import json
```

```python
info = {
    'title': 'Какой-то заголовок',
    'page': 3,
    'tags': ['list1', 'list2'],
    'approved': True
}
```

```python
json_str = json.dumps(info, ensure_ascii=False)
print(f'JSON строка {json_str}')
```

```python
type(json_str)
```

```python
parsed_info = json.loads(json_str)
print(parsed_info)
```

```python
type(parsed_info)
```

```python
info1 = {
    'title': 'Какой-то заголовок',
    'page': 3,
    'tags': ['list1', 'list2'],
    'approved': True
}
```

```python
with open('info.json', 'w', encoding='utf-8') as f:
    json.dump(info1, f, ensure_ascii=False)
```

```python
with open('info1.json', 'w', encoding='utf-8') as f:
    json.dump(info1, f, ensure_ascii=False, indent=4)
```

```python
with open('info1.json', 'r', encoding='utf-8') as f:
    load_inf = json.load(f)
print(load_inf)
```

```python
type(load_inf)
```

```python
with open('data.csv', 'r', newline='', encoding='utf-8') as csv_file:
    reader1 = csv.reader(csv_file, delimiter=',')

    for row in reader1:
        print(row)
```

```python
data_csv = [
    ['Боб1', '125', 'Москва1'],
    ['Аня1', '130', 'Питер1'],
    ['Пётр1', '127', 'Сочи1'],
]
```

```python
with open('data_name.csv', 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file, delimiter=',')
    for row in data_csv:
        writer.writerow(row)
```
