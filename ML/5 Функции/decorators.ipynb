Декоратор — это функция, которая принимает другую функцию и возвращает новую, с добавленным поведением. Часто используется для логирования, аутентификации, измерения времени и т.д.



```python
def decorator(func):
    def wrapper():
        print("Что-то происходит до вызова функции.")
        func()
        print("Что-то происходит после вызова функции.")
    return wrapper

@decorator
def say_hello():
    print("Привет!")

say_hello()
```

@decorator — это то же самое, что say_hello = decorator(say_hello)

```python
import time

def timer(func):
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print(f"Время выполнения: {end - start:.4f} секунд")
    return wrapper

@timer
def slow_function():
    time.sleep(2)
    print("Готово!")

slow_function()
```
