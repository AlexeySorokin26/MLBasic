Генератор — это специальный вид функции, который возвращает значения по одному, а не все сразу, и запоминает, где остановился.

Вместо return в генераторе используется yield.


```python
def my_gen():
    yield 1

for number in my_gen():
    print(number)
```

```python
numbers = []
for i in range(1, 101):
    numbers.append(i)

print(numbers)  # Все числа загружены в память
#Без генератора: Вся последовательность (100 чисел) загружается в память сразу.

def my_gen():
    for i in range(1, 101):
        yield i  # Генератор будет возвращать 1, 2, 3, ..., 100
gen = my_gen()  # Генератор не выполняется, пока не начнём итерировать
print(next(gen))

def my_gen1():
    yield 0
gen1 = my_gen1()
print(next(gen1))
```
