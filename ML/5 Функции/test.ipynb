```python
def make_counter():
    count = 0
    def counter():
        nonlocal count
        count += 1
        return count
    return counter

c = make_counter()
print(c())
print(c())
```

```python
def uppercase(func):
    def wrapper():
        original_res = func()
        modified_res = original_res.upper()
        return modified_res
    return wrapper

@uppercase
def greet():
    return 'Hello!'

print(greet())
        
```
