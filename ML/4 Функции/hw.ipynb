Написать функцию, которая будет перводить снейк_кейс в КэмелКейс и наоборот. Функция сама определяет - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

Пример:

otus_course -> OtusCourse
PythonIsTheBest -> python_is_the_best


```python
def transform_case(word):
    if "_" in word:
        # change into camela
        print("we are in snake_case")
        camela_word = ""
        was__ = False
        for i, c in enumerate(word):
            if i == 0:
                camela_word += c.capitalize()
                continue
            if was__:
                camela_word += c.capitalize()
                was__ = False
            else:
                camela_word += c
            if c == "_":
                was__ = True

        return camela_word
    else:
        # change into snake
        print("we are in CamelaCase")
        SnakeWord = ""
        was_Capital = False
        for i, c in enumerate(word):
            if c.islower() == False:
                if i != 0:
                    SnakeWord += "_"
                SnakeWord += c.lower()
            else:
                SnakeWord += c
        return SnakeWord
                
print(transform_case("otus_course"))
print(transform_case("PythonIsTheBest"))
```

```python
Написать функцию проверяющую валидность введенной даты.

Пример:

29.02.2000 -> True
29.02.2001 -> False
31.04.1962 -> False
```

```python
def check_is_leap_year(year):
    return (year % 4 == 0)
def check_is_month(month):
    return month > 0 and month <= 12
def check_is_day(day, month, is_leap_year):
    if day <= 0 or month < 1 or month > 12:
        return False

    days_in_month = {
        1: 31, 2: 29 if is_leap_year else 28,
        3: 31, 4: 30, 5: 31, 6: 30,
        7: 31, 8: 31, 9: 30, 10: 31,
        11: 30, 12: 31
    }

    return day <= days_in_month[month]
        
# assume user give us numbers 
def check_date(date):
    day, month, year = map(int, date.split("."))
    is_leap_year = check_is_leap_year(year)
    month_correct = check_is_month(month)
    day_correct = check_is_day(day, month, is_leap_year)
    return day_correct and month_correct

print(check_date("29.02.2000"))
print(check_date("29.02.2001"))
print(check_date("31.04.1962"))
print(check_date("31.14.1962"))
```

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

Пример:

17 -> True
20 -> False
23 -> True

```python
def is_prime(n):
    if n == 1 or n == 2:
        return True   
    if n == 0 or n % 2 == 0:
        return False
    for i in range(2, n):  # проверяем все числа от 2 до n-1
        if n % i == 0:
            return False
    return True
print(is_prime(1))
print(is_prime(2))
print(is_prime(3))
print(is_prime(4))
print(is_prime(5))
print(is_prime(123))
print(is_prime(127))
```

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. Также программа должна проверять, что имя и фамилия состоят только из символов и начинаются с большой буквы, если не с большой, то заменяет на большую, возраст должен быть числом от 18 до 60, ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным
Дополнительно написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
def print_table(data):
    print(f"{'ID':<10} {'Name':<15} {'Surname':<15} {'Age':<5}")
    print("-" * 45)
    for user_id, (name, surname, age) in data.items():
        print(f"{user_id:<10} {name:<15} {surname:<15} {age:<5}")

data = {}

while True:
    name = input("Name: ").strip()
    if name == "":
        break
    if not name.isalpha():
        print("Имя должно содержать только буквы.")
        continue
    name = name.capitalize()

    surname = input("Surname: ").strip()
    if surname == "":
        break
    if not surname.isalpha():
        print("Фамилия должна содержать только буквы.")
        continue
    surname = surname.capitalize()

    age_input = input("Age: ").strip()
    if age_input == "":
        break
    if not age_input.isdigit():
        print("Возраст должен быть числом.")
        continue
    age = int(age_input)
    if not (18 <= age <= 60):
        print("Возраст должен быть от 18 до 60.")
        continue

    id_input = input("ID: ").strip()
    if id_input == "":
        break
    if not id_input.isdigit():
        print("ID должен быть числом.")
        continue
    id_formatted = id_input.zfill(8)
    if id_formatted in data:
        print("ID должен быть уникальным.")
        continue

    data[id_formatted] = (name, surname, age)

print_table(data)

    
    
```
