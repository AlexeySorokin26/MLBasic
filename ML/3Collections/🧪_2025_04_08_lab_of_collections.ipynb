# List (—Å–ø–∏—Å–æ–∫)

```python
brain = ["idea", "memory", "memory", "idea", "sidim s bobrom za stolom"]
notebook = brain[0]  # –û–±—Ä–∞—â–µ–Ω–∏–µ –∫ —ç–ª–µ–º–µ–Ω—Ç—É —Å–ø–∏—Å–∫–∞
notebook = brain[-1]  # –û–±—Ä–∞—â–µ–Ω–∏–µ –∫ –ø–æ—Å–ª–µ–¥–Ω–µ–º—É —ç–ª–µ–º–µ–Ω—Ç—É —Å–ø–∏—Å–∫–∞
brain.append("idea")  # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞
brain.extend(["memory", "memory", "memory"]) # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞
brain += ["memory", "memory", "memory"] # –¢–æ –∂–µ —Å–∞–º–æ–µ, –Ω–æ –ø–æ-–¥—Ä—É–≥–æ–º—É
print(brain)
print(f"{len(brain)=}")  # –î–ª–∏–Ω–∞ —Å–ø–∏—Å–∫–∞, f-–Ω–æ—Ç–∞—Ü–∏—è
print("done")
```

üß∂ List comprehension

```python
words = ["–≠—Ç–æ", "–Ø", "–ó–Ω–∞—é", "–ò", "–ü–æ–º–Ω—é", "–ü—Ä–µ–∫—Ä–∞—Å–Ω–æ", "–ü–∏", "–ú–Ω–æ–≥–∏–µ", "–ó–Ω–∞–∫–∏", "–ú–Ω–µ", "–õ–∏—à–Ω–∏", "–ù–∞–ø—Ä–∞—Å–Ω—ã"]
pi_digits = [len(_w) for _w in words]
print(pi_digits)
```

```python
before = [1, 2, 3, 4]
after = [_v+100 for _v in before if _v%2==0]
print(after)
```

```python
before = [1, 2, 3, 4]
after = [_v+100 if _v%2==0 else _v+10000000 for _v in before ]
print(after)
```

üß∂ Dict comprehension

```python
words = ["–≠—Ç–æ", "–Ø", "–ó–Ω–∞—é", "–ò", "–ü–æ–º–Ω—é", "–ü—Ä–µ–∫—Ä–∞—Å–Ω–æ", "–ü–∏", "–ú–Ω–æ–≥–∏–µ", "–ó–Ω–∞–∫–∏", "–ú–Ω–µ", "–õ–∏—à–Ω–∏", "–ù–∞–ø—Ä–∞—Å–Ω—ã"]
pi = {_w: len(_w) for _w in words}
```

```python
products = ["milk", "salt", "tomato"]
costs = [100, 20, 350]
price = {_p: _c for _p, _c in zip(products, costs)}
price
```

## Set (–º–Ω–æ–∂–µ—Å—Ç–≤–æ)
