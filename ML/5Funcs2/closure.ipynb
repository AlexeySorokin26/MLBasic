Замыкание — это функция, которая запоминает переменные из внешней области видимости, даже после того как эта внешняя функция завершила выполнение.



```python
def make_counter():
    count = 0

    def counter():
        nonlocal count
        count += 1
        return count

    return counter

c = make_counter()
print(c())
print(c())
```

Внизу это тоже замыкание 
Тут мы запоминаем function 
Питон это просто все другое

```python
def EnsureVector(function):
    def wrapper(self, other):
        if not isinstance(other, Vector):
            return NotImplemented
        return function(self, other)
    return wrapper

class Vector:
    def __repr__(self):
        return f"Vector{self.coords}"
    
    def __init__(self, *coords):
        self.coords = tuple(c for c in coords)
		
    @EnsureVector
    def __add__(self, other):
        return Vector(*(a+b for a, b in zip(self, other)))
    
    @EnsureVector
    def __sub__(self, other):
        return Vector(*(a-b for a, b in zip(self, other)))

v = Vector(10, 10)
```
