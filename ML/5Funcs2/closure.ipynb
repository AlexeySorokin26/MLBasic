filter(), zip(), map(), range() — все они возвращают итераторы, а не списки.

# 1. map() - итератор
m = map(lambda x: x*2, [1, 2, 3])
print(type(m))  # <class 'map'>
print(list(m))  # [2, 4, 6] - преобразовали в список
print(list(m))  # [] - итератор исчерпан!

# 2. filter() - итератор
f = filter(lambda x: x > 2, [1, 2, 3, 4, 5])
print(type(f))  # <class 'filter'>
print(list(f))  # [3, 4, 5]

# 3. zip() - итератор
z = zip([1, 2, 3], ['a', 'b', 'c'])
print(type(z))  # <class 'zip'>
print(list(z))  # [(1, 'a'), (2, 'b'), (3, 'c')]

# 4. range() - итератор (в Python 3)
r = range(5)
print(type(r))  # <class 'range'>
print(list(r))  # [0, 1, 2, 3, 4]

Замыкание — это функция, которая запоминает переменные из внешней области видимости, даже после того как эта внешняя функция завершила выполнение.
Почему это работает технически:
Когда функция возвращает другую функцию, вместе с ней возвращается всё её "окружение" (все переменные, к которым она обращается). Это называется "лексическое замыкание".

# Посмотрим, что скрыто в функции
def make_counter():
    count = 0
    def counter():
        nonlocal count
        count += 1
        return count
    return counter

c = make_counter()

# У функции есть скрытый атрибут __closure__
print(c.__closure__)  # (<cell at 0x..., int object at 0x...>,)

# В этом замыкании хранится наш count
if c.__closure__:
    cell = c.__closure__[0]
    print(cell.cell_contents)  # 0 - начальное значение count!
    
    c()  # увеличим до 1
    print(cell.cell_contents)  # 1 - изменилось!

```python
def make_counter():
    count = 0

    def counter():
        nonlocal count
        count += 1
        return count

    return counter

c = make_counter()
print(c())
print(c())
```

```python
def EnsureVector(function):
    def wrapper(self, other):
        if not isinstance(other, Vector):
            return NotImplemented
        return function(self, other)
    return wrapper

class Vector:
    def __repr__(self):
        return f"Vector{self.coords}"
    
    def __init__(self, *coords):
        self.coords = tuple(c for c in coords)
		
    @EnsureVector
    def __add__(self, other):
        return Vector(*(a+b for a, b in zip(self, other)))
    
    @EnsureVector
    def __sub__(self, other):
        return Vector(*(a-b for a, b in zip(self, other)))

v = Vector(10, 10)
```
