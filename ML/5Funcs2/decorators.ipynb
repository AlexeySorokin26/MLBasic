Декоратор — это функция, которая принимает другую функцию и возвращает новую, с добавленным поведением. Часто используется для логирования, аутентификации, измерения времени и т.д.



```python
def decorator(func):
    def wrapper():
        print("Что-то происходит до вызова функции.")
        func()
        print("Что-то происходит после вызова функции.")
    return wrapper

@decorator
def say_hello():
    print("Привет!")

say_hello()
```

@decorator def say_hello() — это то же самое, что say_hello = decorator(say_hello)

```python
import time

def timer(func):
    def wrapper():
        start = time.time()
        res = func()
        end = time.time()
        print(f"Время выполнения: {end - start:.4f} секунд")
        return res
    return wrapper

@timer
def do_some_math():
    var = 10 + 10
    return var

print(do_some_math())

def make_timer_great_again(greeting):
    def timer(func):
        def wrapper(extra):
            start = time.time()
            res = func(extra)
            end = time.time()
            print(f"Время выполнения: {end - start:.4f} секунд")
            print(f"some greetings {greeting} for you too {extra}")
            return res
        return wrapper
    return timer

@make_timer_great_again("john")
def do_some_etra_math(extra):
    return 10 + 10

do_some_etra_math("good")
```
