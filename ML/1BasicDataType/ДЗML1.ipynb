```python
days = int(input("Type rest days till vacation:"))
full_weeks = days // 7
weekends = full_weeks * 2
# extra weekends from first week
remaining_days = days % 7
if remaining_days == 6:
  weekends += 1 # saturday
print(weekends)
```

```python
w = int(input("Widht:"))
h = int(input("Height:"))
available_ammount = w * h
bite_size = int(input("Size to bite:"))

# check if we have enough amount
if available_ammount >= bite_size:
# check if we can cut whole bite in one direction (w or h)
  if bite_size % w == 0 or bite_size % h == 0 :
    print("True")
  else:
    print("False")
else:
  print("False")
```

```python
n = int(input("Enter your number:"))
val = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
roms = ['M', 'CM', 'D', 'CD', 'C', 'XC', 'L', 'XL', 'X', 'IX', 'V', 'IV', 'I']

rom_str = ""
for i in range(len(val)):
  count = n // val[i] # how many times we need to repeate rom number
  rom_str += roms[i] * count # repeate current order
  n %= val[i] # decrease our number to one order

print(rom_str)
```

```python
num = input("Enter your num:")
dot_count = 0
only_digits = True
if len(num) == 0:
  print("False")
else:
  for n in num:
    if n == ".":
      dot_count += 1
    elif n < '0' or n > '9':
      only_digits = False
      break
if dot_count <= 1 and only_digits:
  print("True")
else:
  print("False")
```
